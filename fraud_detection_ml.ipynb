```python
# Запустите это в локальном Jupyter Notebook
import subprocess
import sys
import json

def force_install_exact_versions():
    """
    ПРИНУДИТЕЛЬНАЯ установка точных версий из Colab
    """
    # Точные версии из вашего Colab
    exact_versions = {
        'numpy': '2.0.2',
        'pandas': '2.2.2',
        'matplotlib': '3.10.0',
        'seaborn': '0.13.2',
        'scikit-learn': '1.6.1',
        'imbalanced-learn': '0.14.1',
        'lightgbm': '4.6.0',
        'catboost': '1.2.10'
    }
    
    print("="*60)
    print("ПРИНУДИТЕЛЬНАЯ УСТАНОВКА ТОЧНЫХ ВЕРСИЙ")
    print("="*60)
    
    successful = []
    failed = []
    
    for package, version in exact_versions.items():
        package_with_version = f"{package}=={version}"
        
        print(f"\n📦 Устанавливаю: {package_with_version}")
        
        # Сначала удаляем старую версию
        print(f"   🗑 Удаляю старую версию {package}...")
        subprocess.run(
            [sys.executable, "-m", "pip", "uninstall", package, "-y", "--quiet"],
            capture_output=True
        )
        
        # Устанавливаем точную версию
        print(f"   📥 Устанавливаю {package_with_version}...")
        try:
            result = subprocess.run(
                [sys.executable, "-m", "pip", "install", package_with_version, 
                 "--no-cache-dir", "--force-reinstall", "--no-deps"],
                capture_output=True,
                text=True,
                timeout=120
            )
            
            if result.returncode == 0:
                # Устанавливаем зависимости отдельно (важно для некоторых пакетов)
                if package == 'numpy':
                    subprocess.run([sys.executable, "-m", "pip", "install", package_with_version], 
                                 capture_output=True)
                successful.append(package_with_version)
                print(f"   ✅ УСПЕШНО")
            else:
                # Если без зависимостей не получилось, пробуем с зависимостями
                print(f"   ⚠ Пробую с зависимостями...")
                result2 = subprocess.run(
                    [sys.executable, "-m", "pip", "install", package_with_version, 
                     "--force-reinstall"],
                    capture_output=True,
                    text=True,
                    timeout=120
                )
                
                if result2.returncode == 0:
                    successful.append(package_with_version)
                    print(f"   ✅ УСПЕШНО (с зависимостями)")
                else:
                    failed.append(package_with_version)
                    print(f"   ❌ НЕ УДАЛОСЬ")
                    print(f"   Ошибка: {result2.stderr[:200]}")
                    
        except Exception as e:
            failed.append(package_with_version)
            print(f"   ❌ ОШИБКА: {str(e)[:100]}")
    
    # Выводим результаты
    print("\n" + "="*60)
    print("РЕЗУЛЬТАТЫ УСТАНОВКИ")
    print("="*60)
    print(f"✅ Успешно: {len(successful)}")
    print(f"❌ Не удалось: {len(failed)}")
    
    if failed:
        print("\nНеудачные пакеты:")
        for f in failed:
            print(f"   - {f}")
    
    # Проверяем финальные версии
    print("\n" + "="*60)
    print("ФИНАЛЬНАЯ ПРОВЕРКА ВЕРСИЙ")
    print("="*60)
    verify_versions(exact_versions)
    
    return successful, failed

def verify_versions(expected_versions):
    """
    Проверяет, что версии соответствуют ожидаемым
    """
    libs_to_check = {
        'numpy': ('numpy', 'numpy'),
        'pandas': ('pandas', 'pandas'),
        'matplotlib': ('matplotlib', 'matplotlib'),
        'seaborn': ('seaborn', 'seaborn'),
        'scikit-learn': ('sklearn', 'sklearn'),
        'imbalanced-learn': ('imblearn', 'imblearn'),
        'lightgbm': ('lightgbm', 'lightgbm'),
        'catboost': ('catboost', 'catboost')
    }
    
    all_match = True
    
    for lib_name, (import_name, _) in libs_to_check.items():
        try:
            if lib_name == 'scikit-learn':
                import sklearn
                current_version = sklearn.__version__
            elif lib_name == 'imbalanced-learn':
                import imblearn
                current_version = imblearn.__version__
            else:
                module = __import__(import_name)
                current_version = module.__version__
            
            expected_version = expected_versions.get(lib_name)
            
            if current_version == expected_version:
                print(f"✅ {lib_name}: {current_version} (совпадает с Colab)")
            else:
                print(f"⚠️ {lib_name}: {current_version} (должно быть {expected_version})")
                all_match = False
                
        except ImportError:
            print(f"❌ {lib_name}: НЕ УСТАНОВЛЕН")
            all_match = False
    
    if all_match:
        print("\n" + "🎉"*20)
        print("ВСЕ ВЕРСИИ СОВПАДАЮТ С COLAB!")
        print("🎉"*20)
    else:
        print("\n" + "⚠️"*20)
        print("НЕКОТОРЫЕ ВЕРСИИ ОТЛИЧАЮТСЯ")
        print("⚠️"*20)
        print("\nПопробуйте перезапустить ядро и выполнить установку заново")

# Альтернативный метод с использованием conda (если pip не справляется)
def install_with_conda():
    """
    Использует conda для установки точных версий (если установлен conda)
    """
    try:
        # Проверяем наличие conda
        subprocess.run(["conda", "--version"], capture_output=True, check=True)
        
        print("🐍 Использую conda для установки...")
        
        exact_versions = {
            'numpy': '2.0.2',
            'pandas': '2.2.2',
            'matplotlib': '3.10.0',
            'scikit-learn': '1.6.1'
        }
        
        for package, version in exact_versions.items():
            print(f"Устанавливаю {package}=={version} через conda...")
            subprocess.run([
                "conda", "install", "-y", 
                f"{package}={version}", "--quiet"
            ], check=True)
            
        print("✅ Установка через conda завершена")
        
    except (subprocess.CalledProcessError, FileNotFoundError):
        print("⚠️ Conda не найден или не установлен")

# Используем виртуальное окружение для изоляции
def create_isolated_env():
    """
    Создает отдельное виртуальное окружение с точными версиями
    """
    import os
    
    env_name = "colab_env"
    print(f"🔧 Создаю виртуальное окружение '{env_name}'...")
    
    # Создаем окружение
    subprocess.run([
        sys.executable, "-m", "venv", env_name
    ], check=True)
    
    # Определяем путь к pip в окружении
    if sys.platform == "win32":
        pip_path = os.path.join(env_name, "Scripts", "pip")
        python_path = os.path.join(env_name, "Scripts", "python")
    else:
        pip_path = os.path.join(env_name, "bin", "pip")
        python_path = os.path.join(env_name, "bin", "python")
    
    # Устанавливаем пакеты
    packages = [
        "numpy==2.0.2",
        "pandas==2.2.2",
        "matplotlib==3.10.0",
        "seaborn==0.13.2",
        "scikit-learn==1.6.1",
        "imbalanced-learn==0.14.1",
        "lightgbm==4.6.0",
        "catboost==1.2.10"
    ]
    
    for package in packages:
        print(f"Устанавливаю {package}...")
        subprocess.run([pip_path, "install", package], check=True)
    
    print(f"\n✅ Виртуальное окружение создано!")
    print(f"📌 Для активации выполните:")
    if sys.platform == "win32":
        print(f"   {env_name}\\Scripts\\activate")
    else:
        print(f"   source {env_name}/bin/activate")
    print(f"📌 Затем запустите Jupyter из этого окружения")

# ЗАПУСК
print("🚀 НАЧИНАЮ ПРИНУДИТЕЛЬНУЮ УСТАНОВКУ ТОЧНЫХ ВЕРСИЙ\n")

# Выберите метод:
print("Выберите метод установки:")
print("1 - Принудительная установка через pip (рекомендуется)")
print("2 - Создание изолированного виртуального окружения")
print("3 - Использовать conda (если установлен)")

choice = input("\nВаш выбор (1/2/3): ").strip()

if choice == "1":
    force_install_exact_versions()
elif choice == "2":
    create_isolated_env()
elif choice == "3":
    install_with_conda()
    force_install_exact_versions()  # Дополнительно pip для остальных
else:
    print("Неверный выбор, использую метод 1")
    force_install_exact_versions()

print("\n" + "="*60)
print("💡 СОВЕТ: После установки перезапустите ядро Jupyter")
print("💡 Если версии всё ещё не совпадают, выполните:")
print("    !pip install --upgrade pip")
print("    !pip install numpy==2.0.2 --force-reinstall --no-cache-dir")
print("="*60)
```

### FEATURE ENGINEERING

```python
# ==============================================================
# 3. FEATURE ENGINEERING
# ==============================================================

def feature_engineering(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # ── Временные признаки ────────────────────────────────────
    df['trans_datetime'] = pd.to_datetime(df['trans_date_trans_time'])
    df['hour']           = df['trans_datetime'].dt.hour
    df['day_of_week']    = df['trans_datetime'].dt.dayofweek   # 0=пн, 6=вс
    df['month']          = df['trans_datetime'].dt.month
    df['is_weekend']     = (df['day_of_week'] >= 5).astype(int)
    df['is_night']       = ((df['hour'] >= 22) | (df['hour'] <= 5)).astype(int)

    df['dob'] = pd.to_datetime(df['dob'])
    df['age'] = (df['trans_datetime'] - df['dob']).dt.days // 365

    # ── Геопространственные признаки ───────────────────────────
    # Расстояние между владельцем карты и мерчантом (формула Хаверсина)
    R = 6371.0
    lat1 = np.radians(df['lat'])
    lat2 = np.radians(df['merch_lat'])
    dlat = lat2 - lat1
    dlon = np.radians(df['merch_long'] - df['long'])
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    df['distance_km'] = R * 2 * np.arcsin(np.sqrt(a))

    # ── Статистики по клиенту (агрегация по cc_num) ────────────
    client_stats = (df.groupby('cc_num')['amt']
                    .agg(client_mean_amt='mean',
                         client_std_amt='std',
                         client_max_amt='max',
                         client_tx_count='count')
                    .reset_index())
    df = df.merge(client_stats, on='cc_num', how='left')

    # Насколько текущая транзакция выбивается из истории клиента
    df['amt_vs_client_mean'] = df['amt'] / (df['client_mean_amt'] + 1e-6)
    df['amt_zscore']         = ((df['amt'] - df['client_mean_amt'])
                                / (df['client_std_amt'] + 1e-6))
    df['amt_is_max']         = (df['amt'] >= df['client_max_amt']).astype(int)

    # ── Статистики по мерчанту ─────────────────────────────────
    merchant_stats = (df.groupby('merchant')['amt']
                      .agg(merchant_mean_amt='mean',
                           merchant_fraud_rate='mean')  # заглушка, пересчитается ниже
                      .reset_index())
    df = df.merge(merchant_stats, on='merchant', how='left')
    df['amt_vs_merchant_mean'] = df['amt'] / (df['merchant_mean_amt'] + 1e-6)

    # ── Признаки по категории ──────────────────────────────────
    cat_mean = df.groupby('category')['amt'].transform('mean')
    df['amt_vs_category_mean'] = df['amt'] / (cat_mean + 1e-6)

    # ── Логарифм суммы (убирает скошенность) ──────────────────
    df['log_amt'] = np.log1p(df['amt'])

    # ── Признаки из имени мерчанта ────────────────────────────
    df['merchant_name_len'] = df['merchant'].str.len()

    # ── Возрастная группа ─────────────────────────────────────
    df['age_group'] = pd.cut(df['age'],
                             bins=[0, 25, 40, 60, 120],
                             labels=['young', 'adult', 'middle', 'senior'])

    # ── Пол × категория (составной признак) ───────────────────
    df['gender_category'] = df['gender'].astype(str) + '_' + df['category'].astype(str)

    return df


print('Применяем feature engineering...')
train_fe = feature_engineering(train_df)
val_fe  = feature_engineering(val_df)
test_fe  = feature_engineering(test_df)
print(f'Признаков до: {train_df.shape[1]}  |  После: {train_fe.shape[1]}')
```

```python
# ==============================================================
# ПОДГОТОВКА + ОБУЧЕНИЕ
# ==============================================================

TARGET = 'is_fraud'

DROP_COLS = [
    'trans_date_trans_time', 'trans_datetime',
    'cc_num', 'first', 'last', 'street', 'city', 'state', 'zip',
    'trans_num', 'unix_time', 'dob',
    'merchant',
    'job',
    TARGET
]

CAT_COLS = ['category', 'gender', 'age_group', 'gender_category']

# ---------- 1. БАЗОВАЯ ПОДГОТОВКА (без кодирования) ----------
def prepare_base(df, need_ident_cols=False):
    X = df.drop(columns=[c for c in DROP_COLS if c in df.columns], errors='ignore')
    y = df[TARGET].values
    tr_ids = df[['trans_num','trans_date_trans_time']]

    for col in CAT_COLS:
        if col in X.columns:
            X[col] = X[col].astype('category')
            
    if need_ident_cols==True:
        return X, y, tr_ids
    else:
        return X, y

X_train_full, y_train_full = prepare_base(train_fe)
X_val, y_val = prepare_base(val_fe)
X_test, y_test, tr_ids = prepare_base(test_fe, True)

# удаление NaN в таргете
mask_tr = ~np.isnan(y_train_full)
mask_va = ~np.isnan(y_val)
mask_te = ~np.isnan(y_test)

X_train_full = X_train_full[mask_tr]
y_train_full = y_train_full[mask_tr]

X_val = X_val[mask_va]
y_val = y_val[mask_va]

X_test = X_test[mask_te]
y_test = y_test[mask_te]

# сэмпл для "тяжёлых" моделей
X_sample, _, y_sample, _ = train_test_split(
    X_train_full, y_train_full,
    train_size=0.3,
    stratify=y_train_full,
    random_state=SEED
)

num_cols = X_train_full.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in CAT_COLS if c in X_train_full.columns]

# ==============================================================
# 2. LOGISTIC REGRESSION → ONE-HOT + SCALING
# ==============================================================

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.pipeline import Pipeline

preprocessor_lr = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])

X_sample_lr = preprocessor_lr.fit_transform(X_sample)
X_val_lr   = preprocessor_lr.transform(X_val)
X_test_lr   = preprocessor_lr.transform(X_test)

smote = SMOTE(random_state=SEED)
X_sample_lr_sm, y_sample_lr_sm = smote.fit_resample(X_sample_lr, y_sample)

# ==============================================================
# 3. RANDOM FOREST → ORDINAL
# ==============================================================

preprocessor_rf = ColumnTransformer([
    ('num', 'passthrough', num_cols),
    ('cat', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), cat_cols)
])

X_sample_rf = preprocessor_rf.fit_transform(X_sample)
X_val_rf   = preprocessor_rf.transform(X_val)
X_test_rf   = preprocessor_rf.transform(X_test)

X_sample_rf = pd.DataFrame(X_sample_rf).fillna(0)
X_val_rf = pd.DataFrame(X_val_rf).fillna(0)
X_test_rf   = pd.DataFrame(X_test_rf).fillna(0)

smote = SMOTE(random_state=SEED)
X_sample_rf_sm, y_sample_rf_sm = smote.fit_resample(X_sample_rf, y_sample)

# ==============================================================
# 4. LIGHTGBM → CATEGORY dtype
# ==============================================================

cat_features_lgb = [c for c in cat_cols]

# ==============================================================
# 5. CATBOOST → без кодирования
# ==============================================================

cat_features_cb = [X_train_full.columns.get_loc(c) for c in cat_cols]
```

```python
# ==============================================================
# 6. ОБУЧЕНИЕ
# ==============================================================

results = {}

def evaluate(name, model, X_tr, y_tr, X_va, y_va, X_te, y_te, feature_cols):
    print(f'\nОбучение: {name}...')

    t0 = time.time()

    if name == 'CatBoost':
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            cat_features=cat_features_cb,
            verbose=100   # встроенный прогресс
        )

    elif name == 'LightGBM':
        model.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            eval_metric='auc',
            callbacks=[lgb.log_evaluation(100)]  # лог каждые 100 итераций
        )

    else:
        print("   обучение...")
        model.fit(X_tr, y_tr)
        print("   готово")

    elapsed = time.time() - t0

    proba = model.predict_proba(X_te)[:, 1]
    preds = (proba >= 0.5).astype(int)

    roc  = roc_auc_score(y_te, proba)
    pr   = average_precision_score(y_te, proba)
    report = classification_report(y_te, preds, target_names=['Норма', 'Фрод'],
                                   output_dict=True)
    cm = confusion_matrix(y_te, preds)

    # Feature Importance
    feature_importance = None
    if feature_cols is not None:
        if hasattr(model, 'feature_importances_'):
            # Для Random Forest, LightGBM, XGBoost (если не CatBoost)
            importance = model.feature_importances_
            feature_importance = dict(zip(feature_cols, importance))

        elif hasattr(model, 'coef_'):
            # Для логистической регрессии
            importance = np.abs(model.coef_[0])
            feature_importance = dict(zip(feature_cols, importance))

        elif name == 'CatBoost' and hasattr(model, 'get_feature_importance'):
            # Для CatBoost
            importance = model.get_feature_importance()
            feature_importance = dict(zip(feature_cols, importance))

    results[name] = {
        'ROC-AUC': roc,
        'PR-AUC': pr,
        'F1': report['Фрод']['f1-score'],
        'Precision': report['Фрод']['precision'],
        'Recall': report['Фрод']['recall'],
        'Time': elapsed,
        'model': model,
        'proba': proba,
        'preds': preds,
        'classification_report': report,
        'feature_importance': feature_importance
    }

    print(f'   ROC-AUC: {roc:.4f} | PR-AUC: {pr:.4f} | F1: {report["Фрод"]["f1-score"]:.4f} | {elapsed:.1f}s')

# ==============================================================
# 6. ОБУЧЕНИЕ МОДЕЛЕЙ
# ==============================================================
models = [
    ('Logistic Regression',
     LogisticRegression(max_iter=1000, class_weight='balanced',
                        C=0.1, solver='saga', random_state=SEED, n_jobs=-1,verbose=1),
     X_sample_lr_sm, y_sample_lr_sm, X_val, y_val, X_test_lr, y_test),

    ('Random Forest',
     RandomForestClassifier(n_estimators=150, max_depth=12,
                            class_weight='balanced', min_samples_leaf=20,
                            random_state=SEED, n_jobs=-1),
     X_sample_rf_sm, y_sample_rf_sm, X_val, y_val, X_test_rf, y_test),

    ('LightGBM',
     lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05,
                        num_leaves=63,
                        scale_pos_weight=(y_train_full == 0).sum() / (y_train_full == 1).sum(),
                        subsample=0.8, colsample_bytree=0.8,
                        random_state=SEED, n_jobs=-1),
     X_train_full, y_train_full, X_val, y_val, X_test, y_test),

    ('CatBoost',
     CatBoostClassifier(iterations=500, learning_rate=0.05,
                        depth=6, auto_class_weights='Balanced',
                        eval_metric='AUC',
                        random_seed=SEED, verbose=False),
     X_train_full, y_train_full, X_val, y_val, X_test, y_test)
]

feature_cols = X_train_full.columns

for name, model, X_tr, y_tr, X_va, y_va, X_te, y_te in tqdm(models, desc='Общий прогресс'):
    evaluate(name, model, X_tr, y_tr, X_val, y_val, X_te, y_te, feature_cols)

print('\n Все модели обучены!')
```

```python
# =========================================================
# 1. РАСЧЕТ SHAP ДЛЯ ФРОДОВЫХ ТРАНЗАКЦИЙ
# =========================================================

def calculate_shap_for_frauds(model, X_test_frauds):
    """
    Рассчитывает SHAP-значения для фродовых транзакций
    
    Parameters:
    -----------
    model : обученная модель LightGBM
    X_test_frauds : pd.DataFrame - только фродовые транзакции (индекс - tr_id)
    
    Returns:
    --------
    shap_values : np.array - SHAP-значения
    fraud_proba : np.array - вероятности фрода
    tr_ids : list - список ID транзакций
    feature_names : list - названия признаков
    """
    
    # Сохраняем ID транзакций
    tr_ids = X_test_frauds.index.tolist()
    
    # Убираем колонку index, если она есть
    X_for_model = X_test_frauds.copy()
    if 'index' in X_for_model.columns:
        X_for_model = X_for_model.drop(columns=['index'])
    
    feature_names = X_for_model.columns.tolist()
    
    # Создаем explainer
    explainer = shap.TreeExplainer(model)
    
    # Рассчитываем SHAP-значения
    shap_values = explainer.shap_values(X_for_model)
    
    # Для LightGBM shap_values может быть списком [класс_0, класс_1]
    if isinstance(shap_values, list):
        shap_values = shap_values[1]
    
    # Получаем вероятности фрода
    fraud_proba = model.predict_proba(X_for_model)[:, 1]
    
    return shap_values, fraud_proba, tr_ids, feature_names


# =========================================================
# 2. ПРЕОБРАЗОВАНИЕ В LONG-FORMAT С БАКЕТАМИ
# =========================================================

def create_bucket_df_with_transactions(X_test_frauds, shap_values, fraud_proba, tr_ids, feature_names):
    """
    Создает long-format DataFrame с бакетами для каждой транзакции
    
    Parameters:
    -----------
    X_test_frauds : pd.DataFrame - исходные данные (только фичи)
    shap_values : np.array - SHAP-значения (shape: n_samples x n_features)
    fraud_proba : np.array - вероятности фрода
    tr_ids : list - список ID транзакций
    feature_names : list - названия признаков
    
    Returns:
    --------
    result_df : pd.DataFrame с колонками:
        tr_id, fraud_proba, feature, bucket, bucket_left, bucket_right, shap
    """
    
    # Убираем колонку index если она есть
    X_clean = X_test_frauds.copy()
    if 'index' in X_clean.columns:
        X_clean = X_clean.drop(columns=['index'])
    
    # Создаем список для сбора данных
    records = []
    
    # Для каждой транзакции
    for i, tr_id in enumerate(tr_ids):
        
        # Для каждой фичи
        for j, feature in enumerate(feature_names):
            
            feature_value = X_clean.iloc[i, j]
            shap_value = shap_values[i, j]
            
            records.append({
                'tr_id': tr_id,
                'fraud_proba': fraud_proba[i],
                'feature': feature,
                'feature_value': feature_value,
                'shap': shap_value
            })
    
    # Создаем DataFrame
    shap_long = pd.DataFrame(records)
    
    # =========================================================
    # СОЗДАНИЕ БАКЕТОВ
    # =========================================================
    
    def make_bins(x, bins=5):
        """Создает бакеты для числовых и категориальных признаков"""
        x_num = pd.to_numeric(x, errors='coerce')
        
        # Если числовой признак
        if x_num.notna().sum() > 0:
            try:
                # Убираем выбросы для более равномерного распределения
                q_low = x_num.quantile(0.01)
                q_high = x_num.quantile(0.99)
                x_clipped = x_num.clip(q_low, q_high)
                
                return pd.qcut(
                    x_clipped,
                    q=bins,
                    duplicates='drop',
                    labels=False
                )
            except:
                return pd.Series([-1] * len(x), index=x.index)
        else:
            # Категориальный признак
            return x.astype(str)
    
    # Создаем бакеты (группируем по feature)
    shap_long['bucket_idx'] = (
        shap_long.groupby('feature')['feature_value']
        .transform(lambda x: make_bins(x, bins=5))
    )
    
    # Создаем человеко-читаемые бакеты
    def create_bucket_label(group):
        """Создает интервалы для числовых признаков"""
        values = group['feature_value']
        bucket_idx = group['bucket_idx']
        
        # Пытаемся конвертировать в числа
        values_num = pd.to_numeric(values, errors='coerce')
        
        if values_num.notna().sum() > 0:
            # Числовой признак
            buckets = []
            for idx in bucket_idx:
                if idx == -1:
                    buckets.append('single_bin')
                else:
                    # Находим границы для этого бакета
                    mask = bucket_idx == idx
                    if mask.sum() > 0:
                        min_val = values_num[mask].min()
                        max_val = values_num[mask].max()
                        buckets.append(f'({min_val:.2f}, {max_val:.2f}]')
                    else:
                        buckets.append('unknown')
            return pd.Series(buckets, index=group.index)
        else:
            # Категориальный признак
            return values.astype(str)
    
    # Применяем создание бакетов
    shap_long['bucket'] = (
        shap_long.groupby('feature', group_keys=False)
        .apply(create_bucket_label)
        .reset_index(drop=True)
    )
    
    # Извлекаем левую и правую границы для числовых бакетов
    def extract_bucket_bounds(bucket_str):
        """Извлекает левую и правую границу из строки бакета"""
        if bucket_str.startswith('(') and ']' in bucket_str:
            try:
                left_right = bucket_str.strip('()[]').split(',')
                left = float(left_right[0].strip())
                right = float(left_right[1].strip())
                return pd.Series([left, right])
            except:
                return pd.Series([None, None])
        return pd.Series([None, None])
    
    # Добавляем границы
    bounds = shap_long['bucket'].apply(extract_bucket_bounds)
    shap_long['bucket_left'] = bounds[0]
    shap_long['bucket_right'] = bounds[1]
    
    # Убираем временные колонки
    shap_long = shap_long.drop('bucket_idx', axis=1)
    
    return shap_long


# =========================================================
# 3. АГРЕГАЦИЯ (каждая строка = фича в транзакции)
# =========================================================

def aggregate_by_transaction(shap_long_df):
    """
    Агрегирует данные по транзакциям (тот же формат, но с tr_id)
    """
    
    # Группируем по транзакции, фиче и бакету
    result = (
        shap_long_df
        .groupby(
            ['tr_id', 'fraud_proba', 'feature', 'bucket', 'bucket_left', 'bucket_right'],
            dropna=False
        )
        .agg(
            shap=('shap', 'mean'),
            feature_value=('feature_value', 'first'),
            n_obs=('shap', 'size')
        )
        .reset_index()
    )
    
    # Добавляем абсолютный SHAP
    result['abs_shap'] = result['shap'].abs()
    
    # Сортируем внутри каждой транзакции по важности
    result = result.sort_values(
        ['tr_id', 'abs_shap'],
        ascending=[True, False]
    )
    
    return result


# =========================================================
# 4. ОСНОВНАЯ ФУНКЦИЯ (ВСЕ ВМЕСТЕ)
# =========================================================

def generate_fraud_bucket_report(model, X_test_frauds):
    """
    Генерирует полный отчет по фродовым транзакциям
    
    Parameters:
    -----------
    model : обученная модель LightGBM
    X_test_frauds : pd.DataFrame - только фродовые транзакции (индекс - tr_id)
    
    Returns:
    --------
    result : pd.DataFrame - итоговый отчет
        tr_id, fraud_proba, feature, bucket, bucket_left, bucket_right, shap, abs_shap, feature_value
    """
    
    # Шаг 1: Рассчитываем SHAP
    shap_values, fraud_proba, tr_ids, feature_names = calculate_shap_for_frauds(model, X_test_frauds)
    
    # Шаг 2: Создаем long-format с бакетами
    shap_long = create_bucket_df_with_transactions(
        X_test_frauds, 
        shap_values, 
        fraud_proba, 
        tr_ids, 
        feature_names
    )
    
    # Шаг 3: Агрегируем по транзакциям
    result = aggregate_by_transaction(shap_long)
    
    return result

def add_contrib_share(df):
    """
    Добавляет колонку contrib_share = shap / суммарный |shap| по tr_id
    
    Parameters:
    -----------
    df : pd.DataFrame - датафрейм с колонками tr_id и shap
    
    Returns:
    --------
    df : pd.DataFrame - с добавленной колонкой contrib_share
    """
    
    # Копируем чтобы не изменять оригинал
    df = df.copy()
    
    # Считаем суммарный абсолютный SHAP по каждой транзакции
    total_shap_per_tr = df.groupby('tr_id')['abs_shap'].sum()
    
    # Маппим к каждой строке
    df['total_shap_per_tr'] = df['tr_id'].map(total_shap_per_tr)
    
    # Рассчитываем долю вклада (shap / суммарный абсолютный SHAP)
    df['contrib_share'] = df['abs_shap'] / df['total_shap_per_tr']
    
    # Убираем временную колонку
    df = df.drop('total_shap_per_tr', axis=1)
    
    return df

# Запускаем
fraud_bucket_report = generate_fraud_bucket_report(model, X_test_frauds)

fraud_bucket_report = add_contrib_share(fraud_bucket_report)

fraud_bucket_report
```

```python
# =========================================================
# СОЗДАНИЕ БАКЕТОВ
# =========================================================

def make_bins(x, bins=5):

    # пытаемся привести к numeric
    x_num = pd.to_numeric(x, errors='coerce')

    # если получилось много чисел -> qcut
    if x_num.notna().sum() > 0:

        try:
            return pd.qcut(
                x_num,
                q=bins,
                duplicates='drop'
            )

        except:
            return pd.Series(
                ['single_bin'] * len(x),
                index=x.index
            )

    # иначе -> категориальные
    return x.astype(str)

# создаём бакеты
shap_long['bucket'] = (
    shap_long.groupby('feature')['value']
    .transform(make_bins)
)

# =========================================================
# ГРАНИЦЫ ИНТЕРВАЛОВ
# =========================================================

def get_left(x):
    if isinstance(x, pd.Interval):
        return round(x.left, 2)
    return None

def get_right(x):
    if isinstance(x, pd.Interval):
        return round(x.right, 2)
    return None

shap_long['bucket_left'] = shap_long['bucket'].apply(get_left)
shap_long['bucket_right'] = shap_long['bucket'].apply(get_right)

# =========================================================
# АГРЕГАЦИЯ
# =========================================================

shap_agg = (
    shap_long
    .groupby(
        ['feature', 'bucket', 'bucket_left', 'bucket_right'],
        dropna=False
    )
    .agg(
        shap=('shap', 'mean'),
        n_obs=('shap', 'size')
    )
    .reset_index()
)

# абсолютный вклад
shap_agg['abs_shap'] = shap_agg['shap'].abs()

# фильтр мусора
shap_agg = shap_agg[shap_agg['n_obs'] >= 30]

# сортировка
shap_agg = shap_agg.sort_values(
    ['abs_shap', 'n_obs'],
    ascending=[False, False]
)

display(shap_agg.head(30))
```

```python
import pandas as pd
import numpy as np
import shap
from tqdm import tqdm

def daily_shap_analysis_period(model, X_test, fraud_predictions_datamart, start_date='2020-11-01', end_date='2020-11-05', feature_names=None):
    """
    Проводит SHAP-анализ для каждой даты в указанном периоде
    
    Параметры:
    - model: обученная модель
    - X_test: DataFrame с тестовыми данными (фичи)
    - fraud_predictions_datamart: DataFrame с колонками trans_num, tr_date, fraud_pred_proba
    - start_date: начальная дата (включительно)
    - end_date: конечная дата (включительно)
    
    Возвращает:
    - shap_long: длинный фрейм с SHAP значениями для каждой транзакции
    - feature_importance_daily: средние SHAP и доли для каждого признака по дням
    """
    
    # Преобразуем даты к datetime для сравнения
    fraud_predictions_datamart['tr_date'] = pd.to_datetime(fraud_predictions_datamart['tr_date'])
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    # Фильтруем датафрейм по датам (с 1 по 5 ноября 2020)
    date_mask = (fraud_predictions_datamart['tr_date'] >= start_date) & (fraud_predictions_datamart['tr_date'] <= end_date)
    fraud_predictions_filtered = fraud_predictions_datamart[date_mask].copy()
    
    # Фильтруем X_test по тем же индексам
    X_test_filtered = X_test.loc[date_mask[date_mask].index].copy()
    
    if len(X_test_filtered) == 0:
        raise ValueError(f"Нет данных за период с {start_date} по {end_date}")
    
    print(f"Найдено транзакций за период: {len(X_test_filtered)}")
    print(f"Период: с {start_date.date()} по {end_date.date()}")
    
    # Проверяем соответствие
    if len(X_test_filtered) != len(fraud_predictions_filtered):
        raise ValueError(f"Несовпадение размеров после фильтрации: X_test={len(X_test_filtered)}, datamart={len(fraud_predictions_filtered)}")
    
    # Если feature_names не указаны, берем из X_test
    if feature_names is None:
        feature_names = X_test_filtered.columns.tolist()
    
    # Создаем explainer
    if hasattr(model, 'get_booster'):  # XGBoost
        explainer = shap.TreeExplainer(model)
    elif hasattr(model, 'booster_'):  # LightGBM
        explainer = shap.TreeExplainer(model)
    elif hasattr(model, 'tree_') or hasattr(model, 'estimators_'):  # RandomForest
        explainer = shap.TreeExplainer(model)
    else:
        # Для других моделей используем KernelExplainer (медленнее)
        background = X_test_filtered.sample(min(100, len(X_test_filtered)))
        explainer = shap.KernelExplainer(model.predict_proba, background)
    
    # Получаем список уникальных дат в отфильтрованном периоде
    dates = fraud_predictions_filtered['tr_date'].unique()
    dates = sorted(dates)
    print(f"Найдено {len(dates)} уникальных дат в периоде: {[d.strftime('%Y-%m-%d') for d in dates]}")
    
    # Результаты
    all_shap_dfs = []
    daily_importance = []
    
    for date in tqdm(dates, desc="Анализ по дням"):
        # Фильтруем данные за текущую дату
        date_mask_day = fraud_predictions_filtered['tr_date'] == date
        date_indices = date_mask_day[date_mask_day].index
        
        X_test_date = X_test_filtered.loc[date_indices]
        date_transactions = fraud_predictions_filtered.loc[date_indices]
        
        if len(X_test_date) == 0:
            continue
        
        # Рассчитываем SHAP
        try:
            if isinstance(explainer, shap.TreeExplainer):
                shap_values = explainer.shap_values(X_test_date)
                # Если модель бинарная классификации, shap_values может быть списком
                if isinstance(shap_values, list):
                    shap_values = shap_values[1]  # берем для класса 1 (фрод)
            else:
                shap_values = explainer.shap_values(X_test_date)
                if isinstance(shap_values, list):
                    shap_values = shap_values[1]
        except Exception as e:
            print(f"Ошибка SHAP для даты {date.date()}: {e}")
            continue
        
        # Создаем DataFrame с SHAP значениями
        shap_df = pd.DataFrame(
            shap_values,
            columns=[f'shap_{col}' for col in feature_names],
            index=date_indices
        )
        
        # Добавляем метаданные
        shap_df['trans_num'] = date_transactions['trans_num'].values
        shap_df['tr_date'] = date
        shap_df['fraud_pred_proba'] = date_transactions['fraud_pred_proba'].values
        
        # Превращаем в длинный формат
        shap_long_date = shap_df.melt(
            id_vars=['trans_num', 'tr_date', 'fraud_pred_proba'],
            var_name='feature',
            value_name='shap'
        )
        # Убираем префикс 'shap_' из названия фичи
        shap_long_date['feature'] = shap_long_date['feature'].str.replace('shap_', '')
        
        all_shap_dfs.append(shap_long_date)
        
        # Агрегируем для daily_importance
        feature_stats = shap_long_date.groupby('feature')['shap'].agg([
            ('mean_shap', 'mean'),      # средний SHAP (со знаком)
            ('abs_mean_shap', lambda x: x.abs().mean()),  # средний |SHAP|
            ('std_shap', 'std'),        # стандартное отклонение
            ('positive_ratio', lambda x: (x > 0).mean())  # доля положительных
        ]).reset_index()
        
        # Считаем долю вклада (от общей суммы абсолютных SHAP)
        total_abs_shap = feature_stats['abs_mean_shap'].sum()
        if total_abs_shap > 0:
            feature_stats['contribution_pct'] = (feature_stats['abs_mean_shap'] / total_abs_shap) * 100
        else:
            feature_stats['contribution_pct'] = 0
        
        feature_stats['tr_date'] = date
        feature_stats['n_transactions'] = len(X_test_date)
        
        daily_importance.append(feature_stats)
    
    # Объединяем результаты
    shap_long_final = pd.concat(all_shap_dfs, ignore_index=True)
    feature_importance_daily = pd.concat(daily_importance, ignore_index=True)
    
    return shap_long_final, feature_importance_daily, X_test_filtered, fraud_predictions_filtered


# Дополнительная функция: получить фрейм с SHAP вместо исходных значений (только за период)
def get_shap_instead_of_values_period(model, X_test, fraud_predictions_datamart, start_date='2020-11-01', end_date='2020-11-05'):
    """
    Возвращает фрейм, где для каждой транзакции вместо значений фич - их SHAP
    Только за указанный период
    """
    # Фильтруем по датам
    fraud_predictions_datamart['tr_date'] = pd.to_datetime(fraud_predictions_datamart['tr_date'])
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    date_mask = (fraud_predictions_datamart['tr_date'] >= start_date) & (fraud_predictions_datamart['tr_date'] <= end_date)
    fraud_filtered = fraud_predictions_datamart[date_mask]
    X_test_filtered = X_test.loc[date_mask[date_mask].index]
    display(X_test_filtered)
    
    # Создаем explainer
    if hasattr(model, 'get_booster'):
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(X_test_filtered)
        if isinstance(shap_values, list):
            shap_values = shap_values[1]
    else:
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(X_test_filtered)
        if isinstance(shap_values, list):
            shap_values = shap_values[1]
    
    # Создаем фрейм с SHAP значениями
    shap_df = pd.DataFrame(
        shap_values,
        columns=[f'shap_{col}' for col in X_test_filtered.columns],
        index=X_test_filtered.index
    )
    
    # Добавляем идентификаторы и дату
    shap_df['trans_num'] = fraud_filtered['trans_num'].values
    shap_df['tr_date'] = fraud_filtered['tr_date'].values
    shap_df['fraud_pred_proba'] = fraud_filtered['fraud_pred_proba'].values
    
    return shap_df


# Анализ только фродовых транзакций за период (с SHAP > порога)
def analyze_fraud_only_period(model, X_test, fraud_predictions_datamart, start_date='2020-11-01', end_date='2020-11-05', fraud_threshold=0.8):
    """
    Возвращает анализ только по фродовым транзакциям за период
    """
    # Получаем SHAP анализ за период
    shap_long, importance_daily, X_test_period, fraud_period = daily_shap_analysis_period(
        model, X_test, fraud_predictions_datamart, start_date, end_date
    )
    
    # Фильтруем фрод
    fraud_shap = shap_long[shap_long['fraud_pred_proba'] >= fraud_threshold].copy()
    
    if len(fraud_shap) == 0:
        print(f"Нет фродовых транзакций с fraud_pred_proba >= {fraud_threshold}")
        return fraud_shap, None
    
    # Агрегируем по фичам для фрода
    fraud_importance = fraud_shap.groupby('feature').agg({
        'shap': ['mean', 'std', lambda x: (x > 0).mean()],
        'trans_num': 'nunique'
    }).round(4)
    
    fraud_importance.columns = ['mean_shap', 'std_shap', 'positive_ratio', 'n_fraud_transactions']
    fraud_importance = fraud_importance.sort_values('mean_shap', ascending=False)
    
    # Считаем долю вклада для фрода
    total_abs_shap = fraud_shap.groupby('trans_num')['shap'].apply(lambda x: x.abs().sum()).sum()
    fraud_importance['contribution_pct'] = (fraud_importance['mean_shap'].abs() / fraud_importance['mean_shap'].abs().sum()) * 100
    
    return fraud_shap, fraud_importance


# Пример использования
if __name__ == "__main__":
    # Предполагаем, что у вас есть:
    # model - обученная модель
    # X_test - DataFrame с фичами
    # fraud_predictions_datamart - DataFrame с trans_num, tr_date, fraud_pred_proba
    
    # 1. Получаем длинный фрейм SHAP за период с 1 по 5 ноября 2020
    shap_long, importance_daily, X_test_period, fraud_period = daily_shap_analysis_period(
        model, 
        X_test, 
        fraud_predictions_datamart,
        start_date='2020-11-01',
        end_date='2020-11-05'
    )
    
    # 2. Результат 1: фрейм с SHAP для каждой транзакции и фичи
    print("\n" + "="*50)
    print("SHAP по транзакциям (первые 5 строк):")
    print("="*50)
    print(shap_long.head())
    
    print(f"\nВсего записей: {len(shap_long)}")
    print(f"Уникальных транзакций: {shap_long['trans_num'].nunique()}")
    print(f"Диапазон дат: {shap_long['tr_date'].min().date()} - {shap_long['tr_date'].max().date()}")
    
    # 3. Результат 2: важность признаков по дням
    print("\n" + "="*50)
    print("Важность признаков по дням (первые 10 строк):")
    print("="*50)
    print(importance_daily.head(10))
    
    # 4. Анализ только фродовых транзакций
    fraud_threshold = 0.8
    print(f"\n" + "="*50)
    print(f"Анализ фродовых транзакций (fraud_pred_proba >= {fraud_threshold}) за период:")
    print("="*50)
    
    fraud_shap, fraud_importance = analyze_fraud_only_period(
        model, X_test, fraud_predictions_datamart,
        start_date='2020-11-01', end_date='2020-11-05',
        fraud_threshold=fraud_threshold
    )
    
    if fraud_importance is not None:
        print("\nТоп-10 признаков по влиянию на фрод:")
        print(fraud_importance.head(10))
    
    # 5. Детальный анализ по дням для фрода
    print("\n" + "="*50)
    print("Динамика SHAP по дням для топ-3 признаков:")
    print("="*50)
    
    top_features = importance_daily.groupby('feature')['abs_mean_shap'].mean().nlargest(3).index.tolist()
    
    for feature in top_features:
        daily_feature = importance_daily[importance_daily['feature'] == feature]
        print(f"\n{feature}:")
        for _, row in daily_feature.iterrows():
            date_str = row['tr_date'].strftime('%Y-%m-%d')
            print(f"  {date_str}: mean_shap={row['mean_shap']:.4f}, positive_ratio={row['positive_ratio']:.2%}")
    
    # 6. Получаем фрейм с SHAP вместо исходных значений (только за период)
    print("\n" + "="*50)
    print("Фрейм SHAP вместо исходных значений (первые 5 строк):")
    print("="*50)
    shap_instead = get_shap_instead_of_values_period(
        model, X_test, fraud_predictions_datamart,
        start_date='2020-11-01', end_date='2020-11-05'
    )
    print(shap_instead.head())
    
    # 7. Сохраняем результаты
    print("\n" + "="*50)
    print("Сохраняем результаты...")
    print("="*50)
    shap_long.to_csv('shap_by_transaction_20201101_20201105.csv', index=False)
    importance_daily.to_csv('shap_daily_importance_20201101_20201105.csv', index=False)
    if fraud_importance is not None:
        fraud_importance.to_csv('fraud_feature_importance_20201101_20201105.csv')
    
    print("\n✅ Анализ завершен!")
    print(f"📁 Сохранено:")
    print(f"   - shap_by_transaction_20201101_20201105.csv")
    print(f"   - shap_daily_importance_20201101_20201105.csv")
```

```python
import pandas as pd

def build_cdm_layer(df, fraud_threshold=0.5):
    """
    Строит CDM-слой по дням
    """
    # Преобразуем дату
    df['tr_date'] = pd.to_datetime(df['tr_date']).dt.date
    
    # 1. Основные метрики по дням
    result = []
    
    for date in df['tr_date'].unique():
        day_df = df[df['tr_date'] == date]
        
        # Базовые метрики
        day_stats = {
            'tr_date': date,
            'total_transactions': len(day_df),
            'fraud_count': day_df['is_fraud'].sum(),
            'fraud_percent': (day_df['is_fraud'].sum() / len(day_df)) * 100,
            'predicted_fraud_count': (day_df['fraud_pred_proba'] >= fraud_threshold).sum(),
            'mean_amt': day_df['amt'].mean(),
            'median_amt': day_df['amt'].median(),
            'mean_age': day_df['age'].mean(),
            'median_age': day_df['age'].median(),
            'mean_distance_km': day_df['distance_km'].mean(),
            'median_distance_km': day_df['distance_km'].median(),
            'mean_city_pop': day_df['city_pop'].mean(),
            'median_city_pop': day_df['city_pop'].median(),
        }
        
        # 2. Анализ категориальных признаков
        categorical_features = ['category', 'gender', 'hour', 'day_of_week', 'month', 
                                'is_weekend', 'is_night', 'age_group', 'gender_category']
        
        for cat_feat in categorical_features:
            if cat_feat in day_df.columns:
                # Считаем распределение категорий в этот день
                value_counts = day_df[cat_feat].value_counts().to_dict()
                for value, count in value_counts.items():
                    day_stats[f'{cat_feat}={value}'] = count
        
        result.append(day_stats)
    
    cdm_df = pd.DataFrame(result)
    return cdm_df


# Более компактная версия с отдельным фреймом для категориальных признаков
def build_cdm_detailed(df, fraud_threshold=0.5):
    """
    Возвращает два фрейма: основной (метрики) и категориальный (детализация)
    """
    df = df.copy()
    df['tr_date'] = pd.to_datetime(df['tr_date']).dt.date
    df['is_predicted_fraud'] = (df['fraud_pred_proba'] >= fraud_threshold).astype(int)
    
    # Основные метрики
    main_metrics = df.groupby('tr_date').agg(
        total_transactions=('trans_num', 'count'),
        total_unique_cards=('trans_num', 'nunique'),
        fraud_count=('is_fraud', 'sum'),
        fraud_percent=('is_fraud', lambda x: (x.sum() / len(x)) * 100),
        predicted_fraud_count=('is_predicted_fraud', 'sum'),
        predicted_fraud_percent=('is_predicted_fraud', lambda x: (x.sum() / len(x)) * 100),
        mean_amt=('amt', 'mean'),
        median_amt=('amt', 'median'),
        std_amt=('amt', 'std'),
        mean_age=('age', 'mean'),
        median_age=('age', 'median'),
        mean_distance=('distance_km', 'mean'),
        median_distance=('distance_km', 'median'),
        mean_city_pop=('city_pop', 'mean'),
        median_city_pop=('city_pop', 'median'),
        mean_hour=('hour', 'mean'),
        fraud_by_weekend=('is_fraud', lambda x: x[df.loc[x.index, 'is_weekend'] == 1].sum()),
        fraud_by_night=('is_fraud', lambda x: x[df.loc[x.index, 'is_night'] == 1].sum()),
    ).reset_index()
    
    # Категориальные признаки (распределение)
    categorical_features = ['category', 'gender', 'day_of_week', 'month', 'age_group', 'gender_category']
    
    cat_dfs = []
    for cat_feat in categorical_features:
        if cat_feat in df.columns:
            cat_stats = df.groupby(['tr_date', cat_feat]).size().reset_index(name='count')
            cat_stats['feature'] = cat_feat
            cat_stats = cat_stats.rename(columns={cat_feat: 'value'})
            cat_dfs.append(cat_stats)
    
    categorical_metrics = pd.concat(cat_dfs, ignore_index=True) if cat_dfs else pd.DataFrame()
    
    # Процент фрода по категориям
    fraud_by_category = df.groupby(['tr_date', 'category']).agg(
        total=('is_fraud', 'count'),
        fraud=('is_fraud', 'sum')
    ).reset_index()
    fraud_by_category['fraud_rate'] = (fraud_by_category['fraud'] / fraud_by_category['total']) * 100
    
    return main_metrics, categorical_metrics, fraud_by_category


# Использование
main_stats, cat_stats, fraud_by_cat = build_cdm_detailed(fraud_predictions_datamart[fraud_predictions_datamart['tr_date'].isin([
    '2020-11-01',
    '2020-11-02',
    '2020-11-03',
    '2020-11-04',
    '2020-11-05'
])])

print("Основные метрики по дням:")
print(main_stats)

print("\nРаспределение категорий по дням:")
print(cat_stats.head(10))

print("\nДоля фрода по категориям:")
print(fraud_by_cat)
```
